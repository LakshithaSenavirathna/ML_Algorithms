### Decision Tree

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer  # Import the breast cancer dataset from scikit-learn
```

```python
data = load_breast_cancer() # Loading the Dataset
dataset = pd.DataFrame(data=data['data'], columns=data['feature_names']) # Creating a Pandas DataFrame
dataset # Displaying the DataFrame
```

```python
from sklearn.model_selection import train_test_split # Importing train_test_split function

X = dataset.copy() # creating a copy of the dataset
# The .copy() method ensures that any modifications made to x won't affect the original dataset
y = data['target'] # extracting the 'target' column
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
# Splitting Data into Training and Testing Sets
```

```python
from sklearn.tree import DecisionTreeClassifier
# Import the DecisionTreeClassifier class from scikit-learn's tree module


clf = DecisionTreeClassifier(max_depth = 4)
# Create a DecisionTreeClassifier object with default settings
# Here, set the stoping criteria as max_depth = 4 for splitting the decision tree.

clf = clf.fit(X_train, y_train)
# Train the decision tree model using the training data (X_train, y_train)
```

```python
clf.get_params()
# Get the current parameters of the Decision Tree classifier
```

```python
X_test.shape
```

```python
predictions = clf.predict(X_test)
# Predict target values for the test data using the trained model
predictions
```

```python
clf.predict_proba(X_test)[:5]
# Predict class probabilities for the test data
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
# Calculate the accuracy of the model's predictions
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions, labels=[0,1])
# Create confusion matrix to evaluate model performance
```

```python
from sklearn.metrics import precision_score
precision_score(y_test, predictions)
# Calculates the precision of the model's predictions
```

```python
from sklearn.metrics import recall_score
recall_score(y_test, predictions)
# Calculate the recall score of the model's predictions
# Recall = True Positives / (True Positives + False Negatives)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions, target_names = ['malignant', 'benign']))
# Prints a classification report showing model performance metrics
```
